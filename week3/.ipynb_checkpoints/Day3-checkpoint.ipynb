```python
def remove_stopwords(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w.lower() for w in text 
               if w.isalpha() 
               and w.lower() not in stopwords]
    print(content[:50])
    return content

remove_stopwords(news_text)
```

```python
fdist2 = nltk.FreqDist(remove_stopwords(news_text))
fdist2.most_common(50)
```

#5  Investigate the holonym-meronym relations for some nouns. Remember that there are three kinds of holonym-meronym relation, so you need to use: member_meronyms(), part_meronyms(), substance_meronyms(), member_holonyms(), part_holonyms(), and substance_holonyms().

```python
mother=wn.synsets('mother')
house=wn.synsets('house')
love=wn.synsets('love')
mother
```

```python
house
```

```python
love
```

```python
words=[mother,house,love]
for word in words:
    for syn in word:
        print("\nSyn: ", syn)
        print("Definition: ", syn.definition())
        print("member_meronyms: ", syn.member_meronyms())
        print("part_meronyms: ", syn.part_meronyms())
        print("substance_meronyms: ", syn.substance_meronyms())
        print("member_holonyms: ", syn.member_holonyms())
        print("part_holonyms: ", syn.part_holonyms())
        print("substance_holonyms: ", syn.substance_holonyms())
        
```

Ex#13  What percentage of noun synsets have no hyponyms? You can get all noun synsets using wn.all_synsets('n').

```python
all_noun=list(wn.all_synsets('n'))
without_hyponyms=[noun for noun in all_noun if not noun.hyponyms()]
percentage=len(without_hyponyms)*100/len(all_noun)
print("Percentage of noun synsets have no hyponyms: ", percentage)
```

Ex#14  Define a function supergloss(s) that takes a synset s as its argument and returns a string consisting of the concatenation of the definition of s, and the definitions of all the hypernyms and hyponyms of s.

```python
def supergloss(s):
    concat="Synset:\n"+s.definition()+'\nHypernyms:\n'
    for hyper in list(s.hypernyms()):
        concat+=hyper.definition()+'\n'
    concat+="Hyponyms:\n"
    for hypo in list(s.hyponyms()):
        concat+=hypo.definition()+'\n'
    return concat
print(supergloss(wn.synset('car.n.01')))
```
